# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")

query = "SELECT ConvertedComp FROM master;"
data = pd.read_sql_query(query, conn)

data = data.dropna()

plt.figure(figsize=(10, 6))
plt.hist(data['ConvertedComp'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Converted Compensation', fontsize=16)
plt.xlabel('Converted Compensation (USD)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
query = "SELECT Age FROM master;"
data = pd.read_sql_query(query, conn)

data = data.dropna()

plt.figure(figsize=(8, 6))
plt.boxplot(data['Age'], vert=False, patch_artist=True, notch=True)
plt.title('Box Plot of Age', fontsize=16)
plt.xlabel('Age (Years)', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")

query = "SELECT Age, WorkWeekHrs FROM master;"
data = pd.read_sql_query(query, conn)

data = data.dropna()

plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['WorkWeekHrs'], alpha=0.6, edgecolor='k')
plt.title('Scatter Plot of Age vs WorkWeekHrs', fontsize=16)
plt.xlabel('Age (Years)', fontsize=14)
plt.ylabel('Work Week Hours', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master;"
data = pd.read_sql_query(query, conn)


data = data.dropna()

plt.figure(figsize=(12, 8))
bubble_size = data['Age'] / data['Age'].max() * 100

plt.scatter(data['WorkWeekHrs'], data['CodeRevHrs'], s=bubble_size, alpha=0.5, edgecolor='k')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble Size: Age)', fontsize=16)
plt.xlabel('Work Week Hours', fontsize=14)
plt.ylabel('Code Review Hours', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
query = "SELECT DatabaseDesireNextYear FROM DatabaseDesireNextYear;"
data = pd.read_sql_query(query, conn)

databases = data['DatabaseDesireNextYear'].dropna().str.split(';').explode()
database_counts = databases.value_counts()

top_5_databases = database_counts.head(5)

plt.figure(figsize=(10, 8))
plt.pie(
    top_5_databases,
    labels=top_5_databases.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year', fontsize=16)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
query = """
SELECT Age, WorkWeekHrs, CodeRevHrs 
FROM master 
WHERE Age BETWEEN 30 AND 35;
"""
data = pd.read_sql_query(query, conn)

data = data.dropna()

median_data = data.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

median_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35', fontsize=16)
plt.xlabel('Age (Years)', fontsize=14)
plt.ylabel('Hours', fontsize=14)
plt.legend(title='Metrics', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
query = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60;
"""
data = pd.read_sql_query(query, conn)

data = data.dropna()

median_converted_comp = data.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))
median_converted_comp.plot(kind='bar', color='skyblue', edgecolor='k')
plt.title('Median Converted Compensation for Ages 45 to 60', fontsize=16)
plt.xlabel('Age (Years)', fontsize=14)
plt.ylabel('Median Converted Compensation (USD)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
query = "SELECT MainBranch FROM master;"
data = pd.read_sql_query(query, conn)

main_branch_counts = data['MainBranch'].value_counts()

plt.figure(figsize=(10, 8))
main_branch_counts.plot(kind='barh', color='lightgreen', edgecolor='k')
plt.title('Distribution of MainBranch', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Main Branch', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
